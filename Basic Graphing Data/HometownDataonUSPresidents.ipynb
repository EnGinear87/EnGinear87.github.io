```python
"""Import the libraries for pandas and matplotlib to import the excel file and then graph the corresponding data"""

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
dfPUSA = pd.read_csv('presidents.csv')

"""Create a dictionary list of values based on the frequency of presidents born from the given states"""
states = dfPUSA['State']#States
states = dict(dfPUSA.State.value_counts())
lists = states.items() # Items located in the new dictionary
#print(lists) #Verify the list was created successfully

"""Pair the information that was created in the list in x and y values"""
x, y = zip(*lists)

"""Create a graph and start plotting of data"""
fig,ax = plt.subplots()
fig.suptitle('Greatest Number of US Presidents') # Graph title
ax.xaxis.set_label_text('States')   # x-axis caption
plt.xticks(rotation=45)
ax.yaxis.set_label_text('Frequency of Presidents')  # y-axis caption
fig.set_size_inches(20,7)

ax.bar(x,y)
plt.show()
```
